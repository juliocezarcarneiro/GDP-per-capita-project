```python
import pandas as pd

# Step 1: Load the Health Spending per Capita raw data
health_path = "D:/project3/GDP-per-capita-project/Resources/health-expenditure-raw/API_SH.XPD.CHEX.PC.CD_DS2_en_csv_v2_13425.csv"
health_df = pd.read_csv(health_path, skiprows=4)

# Preview the first few rows
health_df.head()
```

```python
# Step 2: Keep 'Country Name', 'Country Code', and years 2002 to 2022
columns_to_keep = ['Country Name', 'Country Code'] + [str(year) for year in range(2002, 2023)]
health_clean_df = health_df[columns_to_keep]

# Preview the cleaned columns
health_clean_df.head()
```

```python
# Step 3: Drop rows with fewer than 14 valid values
health_clean_df = health_clean_df.dropna(thresh=14)

# Show how many countries are left
print("Number of countries after cleaning:", len(health_clean_df))
health_clean_df.head()
```

```python
# Step 4: Save cleaned Health Spending data
output_path = "D:/project3/GDP-per-capita-project/Cleaned_data/health_clean.csv"
health_clean_df.to_csv(output_path, index=False)

print("Cleaned Health Spending data saved to:", output_path)
```
