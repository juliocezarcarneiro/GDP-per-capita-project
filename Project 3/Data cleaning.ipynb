```python
#Dependencies
import pandas as pd
```

```python
# Load Excel files from the 'data' folder
gdp_path = "data/Raw data/gdp-per-capita-worldbank.csv"
healthcare_path = "data/Raw data/annual-healthcare-expenditure-per-capita.csv"
gini_path = "data/Raw data/economic-inequality-gini-index.csv"
income_path = "data/Raw data/world-bank-income-groups.csv"
pop_density_path = "data/Raw data/population-density.csv"

#Create dataframes
gdp_df = pd.read_csv(gdp_path)
healthcare_spending_df = pd.read_csv(healthcare_path)
gini_df = pd.read_csv(gini_path)
income_groups_df = pd.read_csv(income_path)
pop_density_df = pd.read_csv(pop_density_path)
```

```python
#View GDP  raw data
gdp_df.head()
```

```python
#Rename columns
gdp_df.rename(columns={"Entity": "Country",
                      "GDP per capita, PPP (constant 2021 international $)": "GDP per Capita"}, inplace=True)
#preview cleaned data
gdp_df.head()
```

```python
#View healthcare raw data
healthcare_spending_df.head()
```

```python
#Rename columns
healthcare_spending_df.rename(columns={
    "Entity": "Country",
    "Current health expenditure per capita, PPP (current international $)": "Health_Expenditure"
}, inplace=True)

#preview cleaned data
healthcare_spending_df.head()
```

```python
#preview GINI raw data
gini_df.head()
```

```python
#Rename columns
gini_df.rename(columns={
    "Entity": "Country",
    "Gini coefficient": "Gini_Coefficient"
}, inplace=True)

#Drop annotation column
gini_df.drop("990179-annotations", axis =1, inplace =True)

#Preview cleaned data
gini_df.head()
```

```python
#view income groups raw 
income_groups_df.head()
```

```python
#Rename columns
income_groups_df.rename(columns={
    "Entity": "Country",
    "World Bank's income classification": "Income Class"
}, inplace=True)

#preview cleaned data
income_groups_df.head()
```

```python
#View population density per square km raw
pop_density_df.head()
```

```python
#Rename columns
pop_density_df.rename(columns={
    "Entity": "Country",
    "Population density": "Population density per square kilometer"
}, inplace=True)

#Preview cleaned data
pop_density_df.head()
```

```python
# Filter to keep only data from 2000 to 2023
gdp_df = gdp_df[(gdp_df["Year"] >= 2000) & (gdp_df["Year"] <= 2023)]
healthcare_spending_df = healthcare_spending_df[(healthcare_spending_df["Year"] >= 2000) & (healthcare_spending_df["Year"] <= 2023)]
gini_df = gini_df[(gini_df["Year"] >= 2000) & (gini_df["Year"] <= 2023)]
income_groups_df = income_groups_df[(income_groups_df["Year"] >= 2000) & (income_groups_df["Year"] <= 2023)]
pop_density_df = pop_density_df[(pop_density_df["Year"] >= 2000) & (pop_density_df["Year"] <= 2023)]
```

```python
#preview population density data
pop_density_df.head(23)
```

```python
#preview Income group data
income_groups_df.head(23)
```

```python
#Drop Country code columns from all datasets except GDP to help merge data
healthcare_spending_df.drop("Code", axis=1, inplace=True)
gini_df.drop("Code", axis=1, inplace=True)
income_groups_df.drop("Code", axis=1, inplace=True)
pop_density_df.drop("Code", axis=1, inplace=True)
```

```python
# Merge all the datasets
Countries_df = gdp_df.merge(healthcare_spending_df, on=["Country", "Year"], how="outer") \
                  .merge(gini_df, on=["Country", "Year"], how="outer") \
                  .merge(income_groups_df, on=["Country", "Year"], how="outer") \
                  .merge(pop_density_df, on=["Country", "Year"], how="outer")

#Replace NaNs with "No data"
Countries_df.fillna("No data", inplace=True)

#Preview new merged dataframe
Countries_df.head(10)
```

```python
#Preview unquie contries to check if merged properly
unique_countries = Countries_df["Country"].unique()
print(unique_countries)
```

```python
# List of non-country labels to remove
non_countries = [
    'Africa', 'Asia', 'Europe', 'North America', 'Oceania', 'South America',
    'World', 'High-income countries', 'Low-income countries',
    'Middle-income countries', 'Upper-middle-income countries',
    'Lower-middle-income countries', 'Sub-Saharan Africa (WB)',
    'East Asia and Pacific (WB)', 'Latin America and Caribbean (WB)',
    'Middle East and North Africa (WB)', 'North America (WB)',
    'South Asia (WB)', 'Europe and Central Asia (WB)', 'European Union (27)',
    'Argentina (urban)', 'China (rural)', 'China (urban)', 'India (rural)',
    'India (urban)', 'Micronesia (country) (urban)', 'Uruguay (urban)',
]

# Drop rows where 'Country' is in that list
Countries_final_df = Countries_df[~Countries_df["Country"].isin(non_countries)]
```

```python
#Check countries now after removing non-countries
all_countries = Countries_final_df["Country"].unique()
print(all_countries)
```

```python
# Export as csv file
Countries_final_df.to_csv("data/all_countries_final_data.csv", index=False)
```

```python
# Convert to JSON
Countries_final_json = Countries_final_df.to_json(orient="records")

# Save JSON file
with open("data/Countries_final.json", "w") as f:
    f.write(Countries_final_json)
```
