```python
import pandas as pd

# Step 1: Load the GINI Index raw data
gini_path = "D:/project3/GDP-per-capita-project/Resources/gini-index-raw/API_SI.POV.GINI_DS2_en_csv_v2_26341.csv"
gini_df = pd.read_csv(gini_path, skiprows=4)

# Preview the first few rows
gini_df.head()
```

```python
# Step 2: Keep 'Country Name', 'Country Code', and years 2002–2022
columns_to_keep = ['Country Name', 'Country Code'] + [str(year) for year in range(2002, 2023)]
gini_clean_df = gini_df[columns_to_keep]

# Preview the cleaned columns
gini_clean_df.head()
```

```python
# Drop rows with fewer than 14 valid values (keep countries with at least 14 out of 21)
gini_clean_df = gini_clean_df.dropna(thresh=14)

# Show new count
print("Number of countries after adjusting threshold:", len(gini_clean_df))

```

```python
# Step 4: Save cleaned GINI Index data
output_path = "D:/project3/GDP-per-capita-project/Cleaned_data/gini_clean.csv"
gini_clean_df.to_csv(output_path, index=False)

print("Cleaned GINI data saved to:", output_path)
```
