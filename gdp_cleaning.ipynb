```python
import pandas as pd

# Step 1: Load the raw GDP per capita data
gdp_path = "D:/project3/GDP-per-capita-project/Resources/gdp-per-capita-raw/API_NY.GDP.PCAP.CD_DS2_en_csv_v2_26433.csv"
gdp_df = pd.read_csv(gdp_path, skiprows=4)  # Skip the top 4 metadata rows

# Show the first 5 rows
gdp_df.head()
```

```python
# Step 2 : Keep 'Country Name', 'Country Code', and years 2002 to 2022
columns_to_keep = ['Country Name', 'Country Code'] + [str(year) for year in range(2002, 2023)]
gdp_clean_df = gdp_df[columns_to_keep]

# Show the first few rows
gdp_clean_df.head()
```

```python
# Step 3: Drop rows that have too many missing values (e.g., more than 5 missing years)
gdp_clean_df = gdp_clean_df.dropna(thresh=18)  # Keep rows with at least 18 non-NaN values out of 21

# Show how many rows remain
print("Number of countries after cleaning:", len(gdp_clean_df))
gdp_clean_df.head()
```

```python
# Save cleaned GDP data into the Cleaned_data folder
output_path = "D:/project3/GDP-per-capita-project/Cleaned_data/gdp_clean.csv"
gdp_clean_df.to_csv(output_path, index=False)

print("Cleaned GDP data saved to:", output_path)
```
